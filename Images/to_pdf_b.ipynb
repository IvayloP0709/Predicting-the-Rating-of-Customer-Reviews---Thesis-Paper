```python
import matplotlib.pyplot as plt
import numpy as np

# Load confusion matrix images (replace file paths with your own)
confusion_matrix_files = [
    'confusion_matrix_rf_simple.png',
    "confusion_matrix_bert_simple.png",
    "confusion_matrix_roberta_simple.png",
    "confusion_matrix_bert_weight.png",
    "confusion_matrix_rob_weight.png",
    "confusion_matrix_bert_overs.png",
    "confusion_matrix_roberta_overs.png",
    "confusion_matrix_rf_add.png",
    "confusion_matrix_bert_add.png",
    "confusion_matrix_rob_add.png",


]
# Custom titles for each confusion matrix
titles = [
    "Random Forest Simple Confusion Matrix",
    "BERT Simple Confusion Matrix",
    "roBERTa Simple Confusion Matrix",
    "BERT Weighted Confusion Matrix",
    "roBERTa Weighted Confusion Matrix",
    "BERT Oversampled Confusion Matrix",
    "roBERTa Oversampled Confusion Matrix",
    "Random Forest Additional Data Confusion Matrix",
    "BERT Additional Data Confusion Matrix",
    "roBERTa Additional Data Confusion Matrix"
]

# Create a figure and axes for subplots
num_rows = 5
num_cols = 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 12))

# Plot each confusion matrix image on its corresponding subplot
for i, ax in enumerate(axes.flat):
    if i < len(confusion_matrix_files):
        # Load and plot the confusion matrix image without resizing or compression
        img = plt.imread(confusion_matrix_files[i])
        ax.imshow(img, interpolation='nearest')
        ax.set_title(titles[i])  # Set custom title
        ax.axis("off")
    else:
        # Hide unused subplots
        ax.axis("off")

# Adjust subplot layout
plt.tight_layout()

# Display the plot with a dpi of 300
plt.savefig("confusion_matrix_BERT_all.png", dpi=300)

```
